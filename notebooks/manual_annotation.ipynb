```python
from shutil import copyfile
from tqdm import tqdm

target_path = 'all_images'
origin_path = './../datasets/maxar_val/data/'

def copy_all(origin, target):
     imgs = os.listdir(origin)
     for img in tqdm(imgs):
         if img.endswith('.png'):
            copyfile(os.path.join(origin, img), os.path.join(target, img))

copy_all(origin_path, target_path)


```

```python
all_images = os.listdir('all_images')
chunks = 10
nums = [len(all_images)//chunks for _ in range(chunks)]
nums[-1] += len(all_images) % chunks

print('All images are covered: {}'.format(len(all_images) == sum(nums)))
```

```python
origin = 'all_images'
counter = 0
for chunk, num in zip(range(chunks), nums):
    target = 'chunk_{}'.format(chunk+1)
    for i in range(num):
        img = all_images[counter]
        copyfile(os.path.join(origin, img), os.path.join(target, img))

        counter += 1



```
